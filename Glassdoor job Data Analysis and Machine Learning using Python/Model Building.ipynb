```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVR
```

```python
data= pd.read_csv("salary_data_cleaned.csv")
```

```python
data.head()
```

```python
df_model = data[['avg_salary','Rating','Size','Type of ownership','Industry' ,
             'job_state','same_state','age','python_yn','spark','aws','excel','desc_len']]
```

```python
df_model
```

```python
data_ml = pd.get_dummies(df_model , drop_first=True)
```

```python
data_ml
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_ml.drop('avg_salary' , axis = 1)
y = data_ml['avg_salary']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression()
```

```python

lr.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test , y_pred_lr)
```
